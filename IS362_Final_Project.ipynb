The first data was retrieved from [Centers for Disease Control and Prevention](https://wonder.cdc.gov/).
The Natality online database report countsof live births occurring within the United States to U.S. residents and non-residents.  Counts can be obtained by state, county, child's gender and weight, etc.  The data are derived from birth certificates.  For this project, I obtained birth counts by New York county, the year (from 2007 to 2016), and percentage of total births.

The second data I used was retrieved from [The United States Census Bureau](https://www.census.gov/en.html).  I used the income in the past 12 months, year 2016.  The data shows the total number of population for each county and the percentage of income ranges per country.

## Families Household

For this section, I want to compare the income of each counties compared to its birth rates.  I've decided to look more into the "families" household for this analysis.  I will combine income levels in relation to the income classes:

1. **Low Income Class: Below 10,000 to 24,000**

2. **Lower Middle Class: 25,000 - 49,000**

3. **Middle Middle Class: 50,000 - 99,999**

4. **Upper Middle Class: 100,000 or more**

```python
sum_df2 = df2
sum_df2
```

### Families Low Income Data

```python
sum_df2['Families:Low_Income'] = sum_df2['Families:<$10,000'] + sum_df2['Families:$10,000-$14,999'] + \
                                 sum_df2['Families:$15,000-$24,999']
sum_df2['Families:Low_Income']
```

### Families Lower Middle Class Income Data

```python
sum_df2['Families:Lowermiddle_Income'] = sum_df2['Families:$25,000-$34,999'] + sum_df2['Families:$35,000-$49,999']
sum_df2['Families:Lowermiddle_Income']
```

### Families Middle Middle Class Income Data

```python
sum_df2['Families:Middlemiddle_Income'] = sum_df2['Families:$50,000-$74,999'] + sum_df2['Families:$75,000-$99,999']
sum_df2['Families:Middlemiddle_Income']
```

### Families Upper Middle Class Income Data

```python
sum_df2['Families:Uppermiddle_Income'] = sum_df2['Families:$100,000-$149,999'] + sum_df2['Families:$150,000-$199,999'] + \
                                         sum_df2['Families:$200,000_or_more']
sum_df2['Families:Uppermiddle_Income']
```

### Visualizing the Income Class

## Conclusion

Kings County has the highest number of births = 417,865
Tompkins County has the lowest number of births = 2,463

Kings County has the highest population across all household types, while Tompkins is not the lowest (some of the counties that are in the income data is not in the natality data), it is however has one of the lowest population across all household types.

I focused on the household type, "Families", in this project.  Looking at the income data, King's County's income seems to be almost distributed evenly between low income (21.2 percent), low middle class income (20.8 percent), middle middle income (27.4 percent), and upper middle class income (30.8 percent).  Tompkins County, in the other hand, has more middle middle income (38.8 percent) and upper middle class income population (35.8 percent).  The population with low income is only 11.8 percent and the low middle class is 14.4.

Looking at the comparison between these two counties, it seems that the difference in the number of births has something to do with the population's income. The higher the income, the lower the birth rate. I'm not sure why this is so I decided to research this even more.

I found this website, How Income Affects Fertility with reasons why this is the case.

According to a study, across countries, there is a strong negative correlation between GDP and fertility, and within countries there is a strong negative correlation between household income and fertility. Richer countries have lower fertility rates than poor ones, and high-income families in a given country have fewer kids than low-income families do. The data I've analyzed did confirm that this also true in New York counties; evidence that income causes fertility to drop.
